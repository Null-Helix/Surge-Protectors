## Add Cycles

```python
data.insert(0, "cycle", 0)
```

```python
for host in hostnames:
    for device in devices:
        new_data = data.loc[(data['hostName'] == host) & (data['device'] == device)].sort_values(by=['timestamp'])
        num = 1
        for i, x in enumerate(new_data.index):
            try:
                if i != 0 and new_data.loc[new_data.index[i]]['capacity'] - new_data.loc[new_data.index[i-1]]['capacity'] > 2:
                    num += 1
                data.loc[x, 'cycle'] = num
            except:
                data.loc[x, 'cycle'] = num
            
```

## Data Statistics

```python
for host in hostnames:
    HUB = data.loc[(data['hostName'] == host) & (data['device'] == 'HUB')].sort_values(by=['timestamp'])
    print(host + " median: " + str(HUB.temperature.median()) + " mean: " + str(HUB.temperature.mean()))
```

```python
avg_discharge = {}
for host in hostnames:
    for device in ["SPO2SENSOR", "RESPSENSOR"]:
        i = 1
        delta = 0
        discharge = 0
        d = data.loc[(data['hostName'] == host) & (data['device'] == device) & (data["cycle"] == i)].sort_values(by=['timestamp'])
        while d.empty == False:
            discharge += d.tail(1).capacity.iloc[0] - d.head(1).capacity.iloc[0]
            time = (pd.to_datetime(d.tail(1).timestamp.iloc[0]) - pd.to_datetime(d.head(1).timestamp.iloc[0]))
            delta += time.days * 24 + time.seconds / 3600
            i += 1
            d = data.loc[(data['hostName'] == host) & (data['device'] == device) & (data["cycle"] == i)].sort_values(by=['timestamp'])
        if delta != 0:
            avg_discharge[host + "/" + device] = discharge/delta 
        else:
            avg_discharge[host + "/" + device] = 0
```

```python
hoobs = []
sensoor = []
for key in avg_discharge.keys():
    sensoor.append(key[18:])
    hoobs.append(key[:17])
```

```python
discharge = pd.DataFrame(data={'hub': hoobs, 'sensor': sensoor, 'discharge % / hour': avg_discharge.values()})
```
